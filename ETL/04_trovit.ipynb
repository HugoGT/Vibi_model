```python
df = pd.concat([df, trovit], axis=0)
df.reset_index(drop=True, inplace=True)
```

```python
df["year"].fillna(0, inplace=True)
```

```python
for i in df:
    if i == "year":
        continue
    df[i] = df[i].apply(lambda x: x.strip())
```

```python
df.drop_duplicates(inplace=True)
```

```python
checker = []

for i in df["year"]:
    try:
        i = int(i)
    except ValueError:
        i = i.strip()

    checker.append(i)

checker
```

```python
for i in range(len(checker)):
    if not isinstance(checker[i], int):
        checker[i] = 0
    elif checker[i] < 1900:
        checker[i] = 0

df["year"] = checker
df["year"].value_counts()
```

```python
df["area"] = df["area"].apply(lambda x: x.replace(".", "").replace("m²", "").strip())
```

```python
baths = []

for i in df["baths"]:
    try:
        i = int(i)
    except ValueError:
        if not isinstance(i, int):
            if "." in i:
                i = 4
            else:
                i = 2

    baths.append(i)

df["baths"] = baths
df["baths"].value_counts()
```

```python
locations = dict()

for i in df['nearby_locations']:
    item = eval(i)
    for facility in item:
        facility = facility.strip().replace(",", "")
        try:
            locations[facility] += 1
        except KeyError:
            locations[facility] = 1

locations
```

```python
for i in df[df["property_type"] == "Condominio"].index:
    description = df["description"].loc[i].lower()
    if "casa" in description:
        df.at[i, "property_type"] = "Casa"
    else:
        df.at[i, "property_type"] = "Departamento"
```

```python
df['property_type'] = df['property_type'].apply(lambda x: x if x in ['Departamento', 'Casa'] else 'Casa')
```

```python
df["property_type"].value_counts()
```

```python
area = []
for i in df["area"]:
    i = i.replace(",", ".")
    try:
        i = round(float(i))
    except ValueError:
        i = 0

    area.append(i)
```

```python
df["area"] = area
df["area"].value_counts()
```

```python
df[df["area"] < 20].count()
```

```python
df = df[df['area'] >= 20]
```

```python
for i in df.index:
    print(df["location"][i], "\n", df["location_address"][i], "\n")
```

```python
def get_longer_address(row):
    if len(row["location"]) >= len(row["location_address"]):
        return row["location"].strip()
    else:
        return row["location_address"].strip()

df["address"] = df.apply(get_longer_address, axis=1)
```

```python
df = df[df["rooms"] != "1.208.500m²"]
```

```python
rooms = df['rooms'].apply(lambda x: int(x.replace(" Baños", "")) if "Baño" in x else int(x))
```

```python
df["rooms"] = rooms
```

```python
df['price'] = df['price'].apply(lambda x: x.replace(".", ""))
```

```python
price = []

for i in df["price"]:
    if "US$" not in i:
        if "S/" in i:
            i = round(int(i[2:]) / 3.7)
        else:
            i = None
    else:
        i = int(i[3:])

    price.append(i)
```

```python
df["price"] = price
df.dropna(subset=['price'], inplace=True)
```

```python
df.head(2)
```

```python
near_avenue = []
near_cc = []
near_parks = []
near_school = []

for i in df["nearby_locations"]:
    if 'Cerca de avenidas principales' in i:
        near_avenue.append(1)
    else:
        near_avenue.append(0)

    if 'Cerca de centros comerciales' in i:
        near_cc.append(1)
    else:
        near_cc.append(0)

    if 'Cerca de parques' in i:
        near_parks.append(1)
    else:
        near_parks.append(0)

    if 'Cerca de escuelas' in i:
        near_school.append(1)
    else:
        near_school.append(0)
```

```python
df["near_avenue"] = near_avenue
df["near_cc"] = near_cc
df["near_parks"] = near_parks
df["near_school"] = near_school
```

```python
df["security"] = 0
df["elevator"] = 0
df["rest_area"] = 0
df["pool"] = 0
df["parking"] = 0
```

```python
for i in df.index:
    facility = df["facilities"].loc[i].lower() + " " + df["description"].loc[i].lower()
    facility = facility.replace("á", "a").replace("é", "e").replace("í", "i").replace("ó", "o").replace("ú", "u")

    if "cochera" in facility:
        df.at[i, "parking"] = 1

    if "terraza" in facility or "patio" in facility or "balcon" in facility or "area verde" in facility or "areas verdes" in facility or "jardin" in facility or "parrilla" in facility:
        df.at[i, "rest_area"] = 1

    if "jacuzzi" in facility or "piscina" in facility:
        df.at[i, "pool"] = 1

    if "alarma" in facility or "seguridad" in facility or "guardian" in facility or "vigil" in facility or "camara" in facility:
        df.at[i, "security"] = 1

    if "ascensor" in facility or "elevador" in facility:
        df.at[i, "elevator"] = 1
```

```python
columns = ['title', 'address', 'price', 'area', 'baths', 'rooms', 'property_type', 'description', 'year', 'parking', 'near_cc', 'near_school', 'near_parks', 'near_avenue', 'security', 'elevator', 'rest_area', 'pool']

df = df[columns]
df.head(2)
```

```python
df.rename(columns={'baths': 'bath', 'rooms': 'room'}, inplace=True)
```

```python
df.to_csv("dataset/trovit_final.csv", index=False)
```
